```python
with open(r"C:\Users\Lenovo\Desktop\UG Stream\GSE38417_series_matrix.txt", "r") as f:
    geo_lines = f.readlines()
for line in geo_lines[:40]:
    print(line.strip())
```

```python
char_lines = [line for line in geo_lines if line.startswith("!Sample_characteristics_ch1")]
for line in char_lines:
    print(line.strip())
```

```python
import re
char_lines = [
    line.strip().replace("!Sample_characteristics_ch1\t", "")
    for line in geo_lines
    if line.startswith("!Sample_characteristics_ch1")
]
char_split = [line.split("\t") for line in char_lines]
char_df = pd.DataFrame(char_split)
char_df
```

```python
meta_df = char_df.T
meta_df.columns = ["sample_type", "age_raw", "tissue"]
meta_df.head()
```

```python
meta_df["disease_status"] = meta_df["sample_type"].apply(
    lambda x: "Control" if "control" in x.lower() else "DMD"
)
meta_df[["sample_type", "disease_status"]].head(10)
```

Disease status has been annotated to "Control" or "DMD"

```python
def parse_age_to_years(age_str):
    if "unknown" in age_str.lower():
        return None

    years = 0
    months = 0

    y_match = re.search(r"(\d+)\s*yr", age_str)
    if y_match:
        years = int(y_match.group(1))

    m_match = re.search(r"(\d+)\s*mo", age_str)
    if m_match:
        months = int(m_match.group(1))

    return years + months / 12
```

```python
meta_df["age_years"] = meta_df["age_raw"].apply(parse_age_to_years)
meta_df[["age_raw", "age_years"]].head(15)
```

```python
def assign_stage(row):
    if row["disease_status"] == "Control":
        return "Control"
    elif row["age_years"] is None:
        return None
    elif row["age_years"] < 3:
        return "Early"
    elif 3 <= row["age_years"] <= 6:
        return "Mid"
    else:
        return "Late"
```

```python
meta_df["stage"] = meta_df.apply(assign_stage, axis=1)

meta_df[["disease_status", "age_years", "stage"]]
```

```python
start_idx = geo_lines.index("!series_matrix_table_begin\n") + 1
end_idx = geo_lines.index("!series_matrix_table_end\n")

expr_lines = geo_lines[start_idx:end_idx]
from io import StringIO
expr_df = pd.read_csv(
    StringIO("".join(expr_lines)),
    sep="\t"
)

expr_df.head()
```

## Sample Metadata Construction and Alignment

Sample-level metadata were extracted directly from the header annotations of the GEO series matrix file (GSE38417). Since GEO series matrix files store metadata in structured header lines rather than as part of the expression table, metadata parsing was performed independently of expression data loading.

Specifically, `!Sample_characteristics_ch1` fields were used to retrieve:
- Disease status (Control vs DMD)
- Age annotations
- Tissue type

Age values, originally provided as heterogeneous string formats (years, months, or combinations thereof), were converted into numeric values (years) to enable downstream stratification. Based on age, DMD samples were categorized into Early (less than 3 years), Mid (3 to 6 years), and Late (greater than 6 years) disease stages, while control samples were retained as a separate reference group.

At this stage, a clean and reproducible metadata table has been established, forming the foundation for all subsequent differential expression and systems-level analyses.
